```python
# Rounding the integer to the next hundredth value plus an offset of 100
# code used from stackoverflow answer - https://stackoverflow.com/questions/39444665/add-data-labels-to-seaborn-factor-plot
# factorplot has been renamed to catplot
def roundup(x):
    return 100 + int(math.ceil(x / 100.0)) * 100 
```

```python
sns.catplot(x='Sex', data=df, kind='count', alpha=0.7, height=4, aspect=1)

# Get current axis on current figure
ax = plt.gca()

# ylim max value to be set
y_max = df['Sex'].value_counts().max() 
ax.set_ylim([0, roundup(y_max)])

# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='red', ha='center', va='bottom')

plt.show()
```

```python
# explore the genders for different classes

sns.catplot('Pclass', data=df, kind='count', hue='Sex')
# Get current axis on current figure
ax = plt.gca()

# ylim max value to be set
y_max = df['Sex'].value_counts().max() 
ax.set_ylim([0, roundup(y_max)])

# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='red', ha='center', va='bottom')

plt.show()
```

```python
def male_female_child(passenger):
    age, sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
df['Person'] = df[['Age', 'Sex']].apply(male_female_child, axis=1)
```

```python
df['Person'].value_counts()
```

```python
# explore the genders for different classes

sns.catplot('Pclass', data=df, kind='count', hue='Person')
# Get current axis on current figure
ax = plt.gca()

# ylim max value to be set
y_max = df['Sex'].value_counts().max() 
ax.set_ylim([0, roundup(y_max)])

# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='red', ha='center', va='bottom')

plt.show()
```
